```python
import cv2 as c
import numpy as np
import matplotlib.pyplot as plt

img = c.imread("download.jpg")
gray = c.cvtColor(img, c.COLOR_BGR2GRAY)
hist = c.equalizeHist(gray)
threshold_value , otsu =  c.threshold(gray,0,255,c.THRESH_BINARY+c.THRESH_OTSU)
kernel = np.ones((3,3),np.uint8)
opening = c.morphologyEx(otsu,c.MORPH_OPEN,kernel)

plt.figure(figsize=(10, 8))

images = [img, gray, hist, opening]
title=["a","g","h","f"]
for i in range(4):
    plt.subplot(2, 2,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(title[i])
 

plt.show()
```
